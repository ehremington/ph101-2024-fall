```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)

plt.rcParams['text.usetex']= False

with plt.xkcd(scale=.5, randomness=10):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.plot(x,y)
    plt.grid(True, lw=0.5, zorder=0)
    ax.set_xlabel('time (s)')
    ax.set_ylabel(r'acceleration (m/s$^2$)')
    ax.axvline(0, color='k')
    ax.axhline(0, color='k')
    fig.tight_layout()
```

```python
fig.savefig('non-constant-acceleration-xkcd.png', dpi=300)
```

```python
import matplotlib
matplotlib.use("pgf")
plt.rcParams.update({"pgf.texsystem": "pdflatex",
                     'font.family': 'serif',
                     'text.usetex': True,
                     'pgf.rcfonts': False,
                     })
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig1 = plt.figure(figsize=(3,3))
ax1 = fig1.add_subplot()
ax1.plot(x,3*y)
plt.grid(True, lw=0.5)
ax1.set_xlabel('time (s)')
ax1.set_ylabel(r'displacement (meters)')
ax1.axvline(0, color='k')
ax1.axhline(0, color='k')
ax1.set_xticks([1,2,3,4,5,6])
ax1.set_yticks(np.linspace(-3,3,7))
fig1.tight_layout()
```

```python
fig1.savefig('non-constant-acceleration.pgf')
```
