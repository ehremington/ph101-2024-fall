```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,50)
y = np.exp(-(((x-2)/1.1)**2))

with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y,'.')
    plt.grid(True, lw=0.5, zorder=0)
    ax.set_xlabel('time')
    ax.set_ylabel('force (newtons)')
```
